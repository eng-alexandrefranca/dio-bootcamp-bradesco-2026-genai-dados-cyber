## Introdução à automação

## Conceito de automação

Automação é o processo de executar tarefas automaticamente usando programas.

Em vez de executar tarefas manualmente, criamos scripts que realizam essas tarefas de forma automática.

Isso permite:

- economizar tempo
- reduzir erros
- processar grandes volumes de dados

## Cenários reais de automação

A automação é usada em várias áreas:

- geração automática de relatórios
- coleta de dados da internet
- processamento diário de arquivos
- atualização de bancos de dados
- monitoramento de sistemas

Python é muito usado para automação por ser simples e poderoso.

# Manipulação automatizada de arquivos

## Processamento automático de arquivos

Uma das automações mais comuns é processar arquivos automaticamente.

Por exemplo:

- ler arquivos de dados
- calcular métricas
- gerar relatórios

```python
import csv

total = 0

with open("vendas.txt","w") as f:
    f.write("100\n200\n150\n300")
```

```python
with open("vendas.txt","r") as f:

    for linha in f:
        valor = int(linha.strip())
        total += valor

print("Total de vendas:", total)
```

## Integração com banco de dados

Também podemos automatizar o armazenamento de dados em bancos de dados.

```python
import sqlite3

conn = sqlite3.connect("dados.db")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS usuarios(
nome TEXT,
idade INTEGER
)
""")

conn.commit()

print("Tabela criada com sucesso")
```

```python
cursor.execute(
    "INSERT INTO usuarios VALUES (?,?)",
    ("Ana", 25)
)

conn.commit()

print("Usuário inserido no banco de dados")
```

```python
cursor.execute("SELECT * FROM usuarios")

dados = cursor.fetchall()

print("Usuários no banco:")

for usuario in dados:
    print(usuario)
```

```python
conn.close()

print("Conexão com o banco encerrada")
```

## Cron Jobs

Em sistemas Linux utilizamos **cron jobs** para agendar tarefas.

Exemplo:

Executar um script todos os dias às 8h:

0 8 * * * python script.py

```python
import datetime

def gerar_relatorio():

    vendas = [120, 200, 150, 300]

    total = sum(vendas)

    agora = datetime.datetime.now()

    print("Relatório gerado em:", agora)
    print("Total de vendas:", total)

gerar_relatorio()
```

## Como isso funciona na prática

Fluxo real de automação:

1️⃣ criamos um script Python (ex: `relatorio.py`)  
2️⃣ adicionamos um cron job no servidor  

Exemplo de cron job:

0 8 * * * python relatorio.py

3️⃣ o sistema executará o script automaticamente todos os dias às 08:00.

# Projeto prático de automação

## O que é um pipeline de dados?

Um pipeline de dados é um processo automatizado que executa várias etapas:

1. coleta de dados
2. processamento
3. análise
4. geração de resultados

Pipelines são muito usados em:

- análise de dados
- engenharia de dados
- relatórios automáticos

## Dataset real utilizado

Vamos utilizar um dataset real de **vendas de e-commerce**.

Esse dataset contém:

- produtos vendidos
- quantidade
- preço
- país da venda

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df = pd.read_csv(url)

df.head()
```

## Processamento automático dos dados

```python
total_gorjetas = df["tip"].sum()

print("Total de gorjetas:", total_gorjetas)
```

```python
gorjetas_por_dia = df.groupby("day")["tip"].sum()

print(gorjetas_por_dia)
```

## Criando um pipeline automatizado

```python
def pipeline_restaurante():

    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

    df = pd.read_csv(url)

    total_gorjetas = df["tip"].sum()

    relatorio = pd.DataFrame({
        "total_gorjetas":[total_gorjetas]
    })

    relatorio.to_csv("relatorio_gorjetas.csv", index=False)

    print("Pipeline executado com sucesso")
    print("Total de gorjetas:", total_gorjetas)

pipeline_restaurante()
```
