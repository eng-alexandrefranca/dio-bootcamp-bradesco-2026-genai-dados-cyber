# Introdução ao arquivos e dados externos

## O que são dados externos?

Em muitos programas, os dados **não ficam dentro do código**.

Eles ficam em **arquivos externos**, como por exemplo:

- arquivos de texto
- planilhas
- bancos de dados
- APIs na internet

Python consegue:

- ler esses dados
- processar informações
- salvar novos resultados

Isso é essencial para trabalhar com **dados reais**.

Alguns formatos são muito usados em programação:

| Tipo | Uso |

## Função `open()`

Para trabalhar com arquivos em Python usamos a função `open()`.

Estrutura básica:

open("dados.txt", "r")

Os modos mais comuns são:

| Modo | Significado |
|-----|-----|
| r | leitura |
| w | escrita (apaga o conteúdo antigo) |
| a | adicionar conteúdo |

---

## Leitura de arquivos

Para ler um arquivo usamos o modo **"r" (read)**.

Isso permite acessar o conteúdo do arquivo sem modificá-lo.


```python
# Criando um arquivo para teste

with open("dados.txt", "w") as arquivo:
    arquivo.write("Python é uma linguagem poderosa.\n")
    arquivo.write("Estamos aprendendo arquivos.")
```

```python
# Lendo o arquivo completo

with open("dados.txt", "r") as arquivo:
    conteudo = arquivo.read()

print(conteudo)
```

## Função `read()`


- `read()` lê todo o conteúdo do arquivo
- o conteúdo é retornado como texto
- podemos armazenar em uma variável

```python
# Lendo linha por linha

with open("dados.txt", "r") as arquivo:
    for linha in arquivo:
        print(linha.strip())
```

## Função `write()`

Para gravar informações em um arquivo usamos o modo **"w" (write)**.

Esse modo cria o arquivo caso ele não exista.

Se o arquivo já existir, o conteúdo anterior será apagado.

```python
# Escrevendo em um arquivo

with open("relatorio.txt", "w") as arquivo:
    arquivo.write("Relatório de vendas\n")
    arquivo.write("Total: 1500")
```

```python
# Adicionando conteúdo ao arquivo

with open("relatorio.txt", "a") as arquivo:
    arquivo.write("\nNovo registro adicionado.")
```

```python
# Lendo linha por linha

with open("relatorio.txt", "r") as arquivo:
    for linha in arquivo:
        print(linha.strip())
```

## Módulo CSV

Arquivos CSV são usados para representar **tabelas de dados**.

Eles são muito comuns em:

- planilhas
- análise de dados
- exportação de sistemas

Python possui um módulo chamado **csv** para facilitar o trabalho com esse formato.

```python
import csv

# criando um arquivo CSV

dados = [
    ["nome", "idade", "cidade"],
    ["Ana", 25, "São Paulo"],
    ["Carlos", 30, "Rio de Janeiro"],
    ["Maria", 28, "Belo Horizonte"]
]

with open("pessoas.csv", "w", newline="") as arquivo:
    writer = csv.writer(arquivo)
    writer.writerows(dados)
```

```python
# lendo um CSV

import csv

with open("pessoas.csv", "r") as arquivo:
    leitor = csv.reader(arquivo)

    for linha in leitor:
        print(linha)
```

## Módulo JSON

JSON é um formato usado para armazenar e transmitir dados estruturados.

Ele é muito usado em:

- APIs
- aplicações web
- armazenamento de dados

Python possui o módulo **json** para trabalhar com esse formato.

```python
import json

dados = {
    "nome": "Ana",
    "idade": 25,
    "cidade": "São Paulo"
}

with open("usuario.json", "w") as arquivo:
    json.dump(dados, arquivo)
```

```python
# lendo JSON

import json

with open("usuario.json", "r") as arquivo:
    dados = json.load(arquivo)

print(dados)
```

## Integração com banco de dados

Além de arquivos, também podemos armazenar dados em **bancos de dados**.

Um banco de dados permite:

- armazenar grandes quantidades de informação
- organizar dados em tabelas
- consultar dados rapidamente

Em Python podemos usar vários bancos de dados.

Para aprender o conceito, vamos usar **SQLite**, que já vem com Python.

```python
import sqlite3

# criando ou conectando ao banco

conexao = sqlite3.connect("dados.db")

cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE usuarios (
    nome TEXT,
    idade INTEGER
)
""")

conexao.commit()
```

```python
# Inserindo dados no banco

import sqlite3

conexao = sqlite3.connect("dados.db")

cursor = conexao.cursor()

cursor.execute(
    "INSERT INTO usuarios VALUES (?, ?)",
    ("Ana", 25)
)

conexao.commit()
```

```python
# Lendo os dados

import sqlite3

conexao = sqlite3.connect("dados.db")

cursor = conexao.cursor()

cursor.execute("SELECT * FROM usuarios")

dados = cursor.fetchall()

print(dados)
```

# Boas práticas de segurança e integridade

Quando trabalhamos com dados externos, podem ocorrer vários problemas:

- arquivos inexistentes
- dados inválidos
- arquivos corrompidos
- erros de leitura

Por isso, é importante aplicar boas práticas para garantir que o programa seja **seguro e confiável**.

## Exemplo com boas práticas

Agora vamos juntar algumas das práticas:

- verificar se o arquivo existe
- tratar erros
- validar dados

```python
import os

arquivo = "dados.txt"

if os.path.exists(arquivo):

    try:
        with open(arquivo, "r") as f:
            for linha in f:
                valor = linha.strip()

                if valor.isdigit():
                    print("Número válido:", valor)
                else:
                    print("Valor inválido:", valor)

    except Exception as erro:
        print("Erro ao ler arquivo:", erro)

else:
    print("Arquivo não encontrado.")
```

# Automatizando importação e exportação de dados

Em muitos sistemas, os dados precisam ser processados automaticamente.

Por exemplo:

- ler um arquivo de vendas todos os dias
- processar os dados
- gerar um relatório
- salvar os resultados

Esse processo automático é chamado de **automação de dados**.

Muitas empresas utilizam esse tipo de automação para:

- análise de dados
- relatórios automáticos
- integração entre sistemas

```python
import csv
import json

total = 0

# etapa 1: importar dados
with open("pessoas.csv", "r") as arquivo:
    leitor = csv.reader(arquivo)

    next(leitor)

    # etapa 2: processar dados
    for linha in leitor:
        valor = float(linha[1])
        total += valor

# etapa 3: gerar resultado
resultado = {
    "total_vendas": total
}

# etapa 4: salvar relatório
with open("relatorio.json", "w") as arquivo:
    json.dump(resultado, arquivo)

print("Pipeline executado com sucesso.")
```
