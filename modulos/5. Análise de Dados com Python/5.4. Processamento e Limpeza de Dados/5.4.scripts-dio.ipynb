## Introdução à limpeza de dados

## Qualidade de dados

Antes de analisar dados, precisamos garantir que eles possuem **boa qualidade**.

Dados de baixa qualidade podem gerar:

- análises incorretas
- modelos imprecisos
- decisões erradas

Por isso, uma etapa importante da análise de dados é a **limpeza de dados (data cleaning)**.

## Problemas comuns em datasets

Alguns problemas comuns são:

- valores ausentes
- dados duplicados
- valores inconsistentes
- formatos incorretos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

# Tratamento de valores ausentes

## O que é NaN?

NaN significa **Not a Number**.

Ele representa **valores ausentes ou desconhecidos** em um dataset.

```python
df.isnull().sum()
```

## Remoção de valores ausentes

```python
df_limpo = df.dropna()

df_limpo.head()
```

## Preenchimento de valores ausentes

Em vez de remover dados, podemos preencher os valores faltantes.

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"].isnull().sum()
```

# Normalização e padronização

## Transformação de variáveis

Às vezes precisamos transformar dados para facilitar a análise.

```python
df["Age_log"] = df["Age"].apply(lambda x: x)
```

```python
print(df["Age_log"])
```

## Escala de dados

Em alguns casos precisamos colocar os dados na mesma escala.

```python
df["Fare_normalized"] = (df["Fare"] - df["Fare"].min()) / (df["Fare"].max() - df["Fare"].min())

df.head()
```

# Automação de Pré-processamento

## Pipelines de dados

Um pipeline de dados executa várias etapas automaticamente:

1. carregar dados  
2. limpar dados  
3. transformar dados  
4. gerar dataset final

```python
def limpar_dados(df):

    df["Age"].fillna(df["Age"].mean(), inplace=True)

    df["Fare_normalized"] = (
        df["Fare"] - df["Fare"].min()
    ) / (df["Fare"].max() - df["Fare"].min())

    return df
```

```python
df = limpar_dados(df)

df.head()
```
