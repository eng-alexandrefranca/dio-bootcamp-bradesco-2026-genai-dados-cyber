## Introdução às boas práticas

## O que é clean code?

Clean Code significa escrever código que seja:

- fácil de ler
- fácil de entender
- fácil de manter

Um bom código não é apenas funcional, ele também é **organizado e claro**.

👉 "Código é lido mais vezes do que escrito"

## Exemplo ruim vs bom

```python
# código ruim

X = 10
Y = 20
Z = X+Y
print(z)
```

```python
# código melhor

numero1 = 10
numero2 = 20
soma = numero1 + numero2

print(soma)
```

## Regras importantes

- nomes claros
- funções pequenas
- evitar "mágica" no código
- comentar apenas quando necessário

## O que são testes?

Testes garantem que o código está funcionando corretamente.

Eles ajudam a:

- evitar erros
- validar funcionalidades
- facilitar manutenção

## Testes unitários

```python
def soma(a,b):
  return a+b

#teste simples
assert soma(2,3) == 5
```

## O que é refatoração?

Refatorar significa melhorar o código sem alterar seu funcionamento.

```python
def calc(a,b):
  return a+b
```

```python
def calcular_soma(numero1, numero2):
  return numero1+numero2
```

```python
def processar(lista):
  resultado = []
  for i in lista:
    if i > 10:
      resultado.append(i*2)
    return resultado
```

```python
def processar_numeros(lista):
  return [numero * 2 for numero in lista if numero > 10]
```

Melhorias:

- menos código
- mais legível
- mais fácil de manter

## Redução de complexidade

Evite código muito complicado.

Prefira:

- funções simples
- código dividido em partes

```python
# código complexo

def processar(lista):
  return [x*2 for x in lista if x > 10]
```

## O que é performance?

Performance é a velocidade e eficiência do código.

## Profiling

Profiling é analisar o desempenho do código.

```python
import time

inicio = time.time()

for i in range(1000000):
    pass

fim = time.time()

print("Tempo:", fim - inicio)
```

## Melhorando performance

```python
# exemplo com lista
numeros = list(range(1000000))

# forma rápida
soma = sum(numeros)

print(soma)
```
